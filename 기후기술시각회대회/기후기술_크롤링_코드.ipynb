```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time 
import pandas as pd
```

```python
url = 'https://www.ctis.re.kr/ko/techClass/classification.do?key=1141'
driver = webdriver.Chrome(executable_path = 'C:\chromedriver_win32\chromedriver.exe')
driver.get(url)
time.sleep(1)
```

```python
driver.find_element_by_css_selector('#climate_box > div > ul > li:nth-child(1) > button').click()
time.sleep(1)
driver.find_element_by_css_selector('#tc2 > i').click()
driver.find_element_by_css_selector('#tc3 > i').click()
time.sleep(0.5)
driver.find_element_by_css_selector('#mc1_anchor').click()
```

```python
분류=[]
기술이름=[]
기술정의=[]
최고기술수준보유국=[]
우리나라기술수준=[]
격차=[]
j=2

for i in range(2,46): #감축만
    try :
        분류.append(driver.find_element_by_css_selector(f'#mc{j-1}_anchor').text) #분류
        기술이름.append(driver.find_element_by_css_selector(f'#tec{i-1}_anchor').text) #기술이름
        기술정의.append(driver.find_element_by_css_selector('#pdfShows > dl > dd:nth-child(2) > table > tbody > tr > td:nth-child(2)').text) #기술정의
        time.sleep(1)
        최고기술수준보유국.append(driver.find_element_by_css_selector('#pdfShows > dl > dd:nth-child(6) > table > tbody > tr:nth-child(1) > td').text)  #최고 기술 수준 보유국
        우리나라기술수준.append(driver.find_element_by_css_selector('#pdfShows > dl > dd:nth-child(6) > table > tbody > tr:nth-child(2) > td').text) #최고 기술 수준(100%) 대비 우리나라의 기술수준(%)
        격차.append(driver.find_element_by_css_selector('#pdfShows > dl > dd:nth-child(6) > table > tbody > tr:nth-child(3) > td').text)#최고기술보유국과의 격차

        driver.find_element_by_css_selector(f'#tec{i}_anchor').click()
        time.sleep(1)
    except :

        driver.find_element_by_css_selector(f'#mc{j}_anchor').click()
        time.sleep(1)
        j +=1


#         driver.find_element_by_css_selector(f'#tc{k}_anchor').click()
```

```python
df=pd.DataFrame([분류,기술이름,기술정의,최고기술수준보유국,우리나라기술수준,격차]).transpose()
df.columns=['분류','기술이름','기술정의','최고기술수준보유국','최고기술수준(100%)대비 우리나라 기술수준','최고기술보유국과의 격차']
df
```

```python
df.to_csv('기후기술_크롤링.csv',index=False)
```
